```python
from ultralytics import YOLO

model = YOLO('yolov8m.pt')  # better capacity than yolov8s.pt

results = model.train(
    data='/content/ingredient_dataset/data.yaml',
    epochs=30,          # You can increase if needed
    imgsz=640,
    batch=16,           # Lower if memory crashes (try 8)
    patience=10,        # Early stopping if no improvement
    name='ingredient_yolov8_finetuned',
    optimizer='AdamW',  # Better than SGD for fine-tuning
    lr0=0.001,          # Lower LR for fine-tuning
    dropout=0.1,        # Helps avoid overfitting
    pretrained=True,    # Uses pretrained weights
    val=True
)
```
